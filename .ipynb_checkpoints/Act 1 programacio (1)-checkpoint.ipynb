```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
np.set_printoptions(suppress=True)
```

```python
def gradiente(h):
    c=11000
    d=20000
    e=32000
    if 0 <= h < c:
        a = (-0.0065)
        to=288.15
        t=to + (a)*(h-0) 
    elif c <= h < d:
        a = 0
        to = 216.65
        t = to
    elif d <= h <= e:
        a = 0.001
        to=216.65
        t = to + (a)*(h-20000) 
    else:
        a = "Valor erroneo"
        t= "Valor erroneo"
    return(t,to,a)
```

```python
h = input("Inserte la altura")
h = int(h)
t,to,a = gradiente(h)
print(t)
```

```python
rango = np.array([5000,15000,25000])
def gradiente_array(h):
    salida_t = np.array([])
    salida_to = np.array([])
    salida_a = np.array([])
    for i in h:
        c=11000
        d=20000
        e=32000
        h= i
        if 0 <= h < c:
            a = (-0.0065)
            to=288.15
            t=to + (a)*(h-0)
        elif c <= h < d:
            a = 0
            to = 216.65
            t = to
        elif d <= h <= e:
            a = 0.001
            to=216.65
            t = to + (a)*(h-20000)
        else:
            a = np.NaN
            t= np.NaN
        salida_t = np.append(salida_t, t)
        salida_to = np.append(salida_to, to)
        salida_a = np.append(salida_a, a)
    return(salida_t, salida_to, salida_a)
   
salida_t, salida_to, salida_a = gradiente_array(rango)

print(salida_t)
```

```python
def presion_array(h):
    salida_p = np.array([])
    tlist,tolist,alist = gradiente_array(h)
    for i in range(len(tlist)):
        t = float(tlist[i])
        to=float(tolist[i])
        a= float(alist[i])
        if a==-0.0065:
            po = 101325 
            p = po*(t/to)**(-9.81/(a*287))
        elif a==0:
            po=22632
            p = po*(m.e)**((-9.81/(287*t))*(h[i]-11000))
        elif a==0.001:
            po = 5447 
            p = po*(t/to)**(-9.81/(a*287))    
        salida_p = np.append(salida_p, p)
    return(salida_p,tlist, h)
    
presion_array(rango)
```

```python
x = np.arange(0,32001,100)
x
```

```python
p, t, h =presion_array(x)

results = np.vstack((h, t, p)).T

results
```

```python
plt_data = np.vstack((h, t, p))
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(plt_data[0],plt_data[1], c = "red", label = "temperatura")
ax1.set_xlabel('altura')
ax1.set_ylabel('temperatura')
ax2 = ax1.twinx()
ax2.plot(plt_data[0],plt_data[2], c = "blue", label = "presion")
ax2.set_ylabel('presion')
fig.legend(loc='upper right')
plt.show()
```
